A scatterplot for continuous data

```python
list_numbers = [1, 2, 3, 4]
list_numbers_sqr = [1, 4, 9, 16]
print(list(list_numbers))
print(list_numbers_sqr)
```

```python
type(list_numbers)
```

```python
plt.scatter(x = list_numbers, y = list_numbers_sqr, color = "red")
plt.xlabel("A meaningful name for the X-axis")
plt.ylabel("Favourite name for Y-axis")
plt.show()

plt.scatter(x = some_2021_track_ppl, y = list_numbers_sqr)
plt.show()
```
